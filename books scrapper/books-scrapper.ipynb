```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
from tqdm import tqdm
```

```python
books = []

for i in tqdm(range(1,51)):

    link = 'https://books.toscrape.com/catalogue/page-'+str(i)+'.html'
    res = requests.get(link)
    soup = BeautifulSoup(res.text)

    for sp in soup.find_all('article', class_ = 'product_pod'):

        name = sp.find('h3').find('a').get('title')
        link = 'https://books.toscrape.com/catalogue/' + sp.find('h3').find('a').get('href')
        
        res = requests.get(link)
        sup = BeautifulSoup(res.text)
        
        book_type = sup.find('ul', class_ = 'breadcrumb').find_all('a')[2].text
        table = sup.find('table', class_ = 'table table-striped').find_all('td')
        UPC = table[0].text
        price_incl = table[2].text[2:]
        price_excl = table[3].text[2:]
        tax = table[4].text[2:]
        stock = table[5].text.split()[-2][1:]
        reviews = table[6].text

        rating = sp.find('p').get('class')[-1]
        price = sp.find('p', class_ = 'price_color').text[2:]
        availability = sp.find('p', class_ = 'instock availability').text.strip()

        books.append([name, book_type, rating, price, UPC, price_incl, price_excl, tax, availability, stock, reviews, link])
```

```python
data = pd.DataFrame(books, columns = ['Name', 'Category', 'Rating', 'Price', 'UPC', 'Price (incl. tax)', 'Price (excl. tax)', 'Tax', 'Stock availability', 'Stock', 'No. of Reviews', 'Link'])
data
```

```python
data.to_csv('books.csv', index = False)
```
