# Importing libraries

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from tqdm import tqdm
```

# Scrapping multiple pages

```python
quotes = []

for page in tqdm(range(1,11)):
    
    link = 'https://quotes.toscrape.com/page/' + str(page)
    res = requests.get(link)
    soup = BeautifulSoup(res.text, 'html.parser')

    for qt in soup.find_all('div', class_ = 'quote'):
        
        quote = qt.find('span', class_ = 'text').text[1:-1]
        author = qt.find('small', class_ = 'author').text
        author_id = qt.find('a').get('href')
        author_link = 'http://quotes.toscrape.com/' + author_id
        tags = []
        for tag in qt.find_all('a', class_ = 'tag'):
            tags.append(tag.text)
        tags = ",".join(tags)
        
        quotes.append([quote, author, author_id, tags, author_link])
```

```python
len(quotes)
```

```python
data = pd.DataFrame(quotes, columns = ['quote','author','author_id','tags', 'author_link'])
```

```python
data.head()
```

```python
data.to_csv('quotes.csv', index = False)
```

# Scrapping author's info

```python
df = data.drop(columns = ['quote', 'author_id', 'tags'])
df = df.drop_duplicates()
df
```

```python
len(df)
```

```python
for author in tqdm(df.values):
    file_name = 'Authors/' + author[0] + '.txt'
    link = author[1]
    
    res = requests.get(link)
    soup = BeautifulSoup(res.text, 'html.parser')
    
    text = soup.find('div', class_ = 'author-details').text
        
    fd = open(file_name, 'w', encoding="utf-8")
    fd.write(text)
    fd.close()
```
