```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
driver.get(link)
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
data = []

for sp in soup.find_all('ytd-rich-grid-media'):
    
    try:
        title = sp.find('yt-formatted-string', {"id": "video-title"}).text
    except:
        title = np.nan
    
    try:
        video_link = sp.find('a', {"id": "video-title-link"}).get('href')
    except:
        video_link = np.nan
    
    try:
        video_duration = sp.find_all('span')[0].text.strip()
    except:
        video_duration = np.nan
        
    try:
        views = sp.find_all('span')[3].text
    except:
        views = np.nan
    
    try:
        date_time = sp.find_all('span')[4].text
    except:
        date_time = np.nan
        
    try:
        thumbnail_link = sp.find('img', {"class":"yt-core-image--fill-parent-height yt-core-image--fill-parent-width yt-core-image yt-core-image--content-mode-scale-aspect-fill yt-core-image--loaded"}).get('src').split('?')[0]
    except:
        thumbnail_link = np.nan
        
    
    data.append([title, video_link, video_duration, views, date_time, thumbnail_link])
```

```python
data
```

```python
df = pd.DataFrame(data, columns = ['title', 'video_link', 'video_duration', 'views', 'date_time', 'thumbnail_link'])
```

```python
df
```

```python
df.to_csv('yt-gfg-data.csv', index = False)
```

# Scrapping data from each video page

```python
data = []

for link in tqdm(df['video_link']):
    
    link = 'https://www.youtube.com' + link
    driver.get(link)
    
    time.sleep(5)
    
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    title = soup.find('h1', {"class":"style-scope ytd-watch-metadata"}).text.strip()
    likes = soup.find('ytd-segmented-like-dislike-button-renderer').find('span', {"class":"yt-core-attributed-string yt-core-attributed-string--white-space-no-wrap"}).text
    views = soup.find('div', {'id':'bottom-row'}).find_all('span')[0].text
    date_time = soup.find('div', {'id':'bottom-row'}).find_all('span')[2].text
    description = soup.find('div', {'id':'description'}).find('span', {"class":"style-scope yt-formatted-string"}).text
    
    data.append([title, likes, views, date_time, description])
```
